```python
import pandas as pd

file_path_incendie = "../data/incendies/incendies.csv"
file_path_INSEE = "../data/base-officielle-codes-postaux.csv"

# Charger le fichier téléchargé sur data.gouv.fr
df_incendie = pd.read_csv(file_path_incendie)
df_INSEE = pd.read_csv(file_path_INSEE,usecols=['code_commune_insee', 'latitude', 'longitude']).drop_duplicates(subset='code_commune_insee', keep='first')

df_merged = df_incendie.merge(
    df_INSEE,
    left_on='Code INSEE',      # colonne dans df_incendie
    right_on='code_commune_insee',  # colonne dans df_geo
    how='left'                 # garde toutes les lignes de df_incendie
)

# Supprimer la colonne redondante 'code_commune_insee' si tu veux
df_merged = df_merged.drop(columns=['code_commune_insee'])
```

```python
# Enregistrer le DataFrame fusionné dans un fichier CSV
output_file = "../data/incendies/incendies_V2.csv"
df_merged.to_csv(output_file, index=False)

print(f"Le fichier a été enregistré sous : {output_file}")
```
